Notebook wrapper untuk menjalankan LightGBM/XGBoost/CatBoost dan sampled SHAP explainability.


## Prinsip

- Bandingkan advanced model dengan baseline Fase 4 pada track yang sama.
- Accuracy bukan klaim utama; prioritaskan macro F1, MCC, recall, dan FP/FN.
- SHAP sampled dipakai sebagai evidence interpretasi forensik, bukan bukti kausal absolut.


```python
from pathlib import Path
import subprocess, sys
cwd = Path.cwd().resolve()
ROOT = cwd if (cwd / 'scripts' / 'run_advanced_modeling.py').exists() else cwd.parent
assert (ROOT / 'scripts' / 'run_advanced_modeling.py').exists(), f'Cannot find repo root from {cwd}'
cmd = [sys.executable, str(ROOT / 'scripts' / 'run_advanced_modeling.py'), '--models', 'all', '--tracks', 'all', '--shap-sample', '3000']
print('Running:', ' '.join(cmd))
subprocess.run(cmd, check=True, cwd=ROOT)
subprocess.run([sys.executable, str(ROOT / 'scripts' / 'generate_dashboard_data.py')], check=True, cwd=ROOT)
```
